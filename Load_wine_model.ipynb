```python
import sklearn as klearn
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

```python
dataset = load_wine()
x_data = dataset.data
y_labels  = dataset.target
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x_data, y_labels, test_size = 0.2, shuffle = True, random_state = 12345)
```

```python
xTrain = xTrain.astype("float32")
xTest = xTest.astype("float32")
```

```python
import tensorflow as tf
import keras
from keras.layers import Dense
from keras import Sequential
```

```python
model = Sequential()
model.add(Dense(xTrain.shape[0], activation = "relu"))
model.add(Dense(xTrain.shape[0]*2, activation = "relu"))
model.add(Dense(3, activation = "softmax"))
```

```python
model.compile(loss = "sparse_categorical_crossentropy", optimizer = "rmsprop", metrics = ["accuracy"])
```

```python
model.fit(xTrain, yTrain, batch_size = 20, epochs = 50)
```

```python
model.evaluate(xTest, yTest, batch_size =10)
```
